```python
# import necessary things
import numpy as np
import matplotlib.pyplot as plt


#assumed values of m,c
m = 2
c = 5

#number of datapoints 
n = 100  

# Generate random x values
x = np.linspace(0, 10, n)

for i in range(3):

    # Generate random error
    e = np.random.normal(i, i+1, n)  # mean = 0, std deviation = 1 initially

    # Calculate y values with added noise
    y2 = m * x + c + e

    # calculation for the farmula for the calculation of the predicted values of m and c
    # average of randomaly genarated x values
    X=(sum(x))/len(x)

    # Average of y values with noise
    Y=(sum(y2))/len(y2)

    product=x*y2
    product_avg=(sum(product)/len(product))
    squa_avg=x*x
    X2=(sum(squa_avg)/len(squa_avg))

    # predicted values of the free parameters
    m_new=(product_avg-X*Y)/((X2)-(X**2))
    c_new=Y-(m_new*X)

    print("predicted Value of m ",m_new)
    print("Predicted value of n ",c_new)
    #Plot the synthetic data
    plt.scatter(x, y2, color='blue', label='Synthetic Data')
    plt.plot(x, m * x + c, color='red', label='True Line (without error)')
    plt.plot(x,m_new * x + c_new,color='green',label='Predicted line')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
```
